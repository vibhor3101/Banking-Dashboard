```python
#Numerical Analysis
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance', 'Bank Loans',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
       'Foreign Currency Account']

#Univate Analysis and visualization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,2,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)

plt.tight_layout(pad=2.0)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

## Heat Maps


```python
numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance', 'Bank Loans',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
       'Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5,fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### Insights from Heat Map


1. Strong Positive Relationships
Bank Deposits ↔ Checking Accounts (r = 0.84)

Bank Deposits ↔ Saving Accounts (r = 0.75)

📌 Interpretation:
Customers who have high bank deposits are very likely to maintain both checking and saving accounts, indicating bundled product usage and disciplined banking behavior.

2. Moderate Positive Correlations
Saving Accounts ↔ Checking Accounts (r = 0.46)

Bank Loans ↔ Business Lending (r = 0.42)

Estimated Income ↔ Superannuation Savings (r = 0.37)

Estimated Income ↔ Bank Loans (r = 0.33)

Credit Card Balance ↔ Bank Deposits (r = 0.38)

📌 Interpretation:

Customers with higher income tend to have more savings and loans, but not extremely so.

Business lending often coexists with personal or corporate loans, suggesting diversified credit usage.

❌ 3. Weak or No Correlation
Estimated Income ↔ Saving Accounts (r = 0.18)

Superannuation Savings ↔ Checking Accounts (r = 0.20)

Foreign Currency Account has low correlation with most variables

📌 Interpretation:

Customers’ income is not a strong predictor of how much they maintain in basic accounts or foreign currency holdings.

Likely influenced more by behavioral or demographic factors than just income level.
